# Overfitting vs. Underfitting

```
import numpy as np
import pandas as pd

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.titlesize'] = 16
matplotlib.rcParams['figure.figsize'] = [9,6]
```

## Generate the data

```
np.random.seed(42)

def true_gen(x):
  y = np.sin(1.2 * x * np.pi)
  return (y)

x = np.sort(np.random.rand(30))
y = true_gen(x) + 0.1*np.random.randn(len(x))
```

```
plt.style.use('seaborn-white')
fig,ax = plt.subplots()
ax.plot(x,y,'ro',label='Training data')

ax.legend(loc='best',frameon=True,edgecolor='black')
plt.show()
```

```
features_1 = PolynomialFeatures(degree = 1, include_bias=False)
features_2 = PolynomialFeatures(degree = 4, include_bias=False)
features_3 = PolynomialFeatures(degree = 35, include_bias=False)
```

```
x = x.reshape((-1,1))
x_trans_1 = features_1.fit_transform(x)
x_trans_2 = features_2.fit_transform(x)
x_trans_3 = features_3.fit_transform(x)
```

```
model = LinearRegression()
model.fit(x_trans_1,y)
train_predictions_1 = model.predict(x_trans_1)
model.fit(x_trans_2,y)
train_predictions_2 = model.predict(x_trans_2)
model.fit(x_trans_3,y)
train_predictions_3 = model.predict(x_trans_3)
```

```
plt.plot(x[:,0,],y,'ro',label='Training Data')
plt.plot(x[:,0],train_predictions_1,'b',linewidth=4,label='Underfitting')
plt.legend(frameon=True,edgecolor='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Underfitting')
```

```
plt.plot(x[:,0,],y,'ro',label='Training Data')
plt.plot(x[:,0],train_predictions_2,'b',linewidth=4,label='Balanced')
plt.legend(frameon=True,edgecolor='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Balanced')
```

```
plt.plot(x[:,0,],y,'ro',label='Training Data')
plt.plot(x[:,0],train_predictions_3,'b',linewidth=4,label='Overfitting')
plt.legend(frameon=True,edgecolor='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Overfitting')
```
